```python
import numpy as np
import struct

def read_idx3(filename):
    with open(filename, 'rb') as f:
        # Read magic number (4 bytes, MSB first)
        magic, num_images = struct.unpack('>II', f.read(8))
        rows, cols = struct.unpack('>II', f.read(8))
        
        # Read image data (unsigned bytes)
        data = np.frombuffer(f.read(), dtype=np.uint8)
        data = data.reshape(num_images, rows, cols)
        
        return data
```

```python
# Load the IDX3 file (e.g., MNIST images)
images = read_idx3('C:\\Users\\nissa\\Downloads\\train-images-idx3-ubyte\\train-images.idx3-ubyte')

# Check shape (num_images, height, width)
print(images.shape)  # e.g., (60000, 28, 28) for MNIST

# Display the first image
import matplotlib.pyplot as plt
plt.imshow(images[0], cmap='gray')
plt.show()
```

```python
# Normalize pixel values to [0, 1]
images_normalized = images / 255.0
```

```python
plt.imshow(images_normalized[0], cmap='gray')
plt.show()
```

```python
X = images_normalized.reshape(images_normalized.shape[0], -1).T
```

```python
classes = 10
W = np.random.randn(X.shape[0], classes)*.01
b = np.zeros((classes, 1))
```
